```python
## Example queries for song play analysis
```

```python
#1.  Count of paid users for a given location
```

```python
%sql select location, count(user_id) as cnt from songplays where level='paid' group by location order by cnt desc;
```

```python
#2.  Count of free users for a given location
```

```python
%sql select location, count(user_id) as cnt from songplays where level='free' group by location order by cnt desc;
```

```python
#3. The total number of free and paid sessions of a user of the app:
```

```python
%sql select songplays.user_id, users.first_name, users.last_name, users.gender, users.level,  count(session_id) cnt from songplays, users where songplays.user_id=users.user_id  group by songplays.user_id, users.first_name, users.last_name, users.gender, users.level order by songplays.user_id;
```

```python
# 4. Most common device platforms of app usage
```

```python
%sql select user_agent, count(session_id) as cnt from songplays group by user_agent order by cnt desc;
```
